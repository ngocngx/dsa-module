```python
class Stack:
    """LIFO Stack implementation using a Python list as underlying storage."""

    def __init__(self):
        self.stack = []
    
    def __len__(self):
        return len(self.stack)

    def __str__(self):
        return str(self.stack)

    def push(self, value):
        self.stack.append(value)
    
    def pop(self):
        if self.is_empty():
            raise Empty('Stack is empty')
        return self.stack.pop()

    def top(self):
        if self.is_empty():
            raise Empty('Stack is empty')
        return self.stack[-1]

    def is_empty(self):
        return len(self) == 0
    
    def move(self, des):
        aux = Stack()

        for i in range(len(self)):
            aux.push(self.pop())

        for i in range(len(aux)):
            des.push(aux.pop())

        return des
    
    def sort(self):
        temp = Stack()
        temp.push(self.pop())

        while self and temp:
            pivot = self.pop()

            # while temp and temp.top() < pivot:
            #     self.push(temp.pop())
            # temp.push(pivot)

            # while temp:
            #     self.push(temp.pop())

            if pivot < temp.top():
                temp.push(pivot)
            else:
                while temp:
                    self.push(temp.pop())
                self.push(pivot)

        return self

A = Stack()
for num in (4, 1, 2, 3):
    A.push(num)

B = Stack()
B = A.move(B)
print(B.sort())
```
